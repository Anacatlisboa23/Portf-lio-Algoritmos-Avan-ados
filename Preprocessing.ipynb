```python
# Load ProtBERT model and tokenizer
protbert_model = BertModel.from_pretrained('Rostlab/prot_bert', return_dict=True)
tokenizer = BertTokenizer.from_pretrained('Rostlab/prot_bert')
```

```python

# Function to analyze protein sequence
def analyze_protein_sequence(sequence,max_length=100):
    # Tokenize the protein sequence
    tokens = tokenizer.encode(sequence, return_tensors='pt', truncation=True)

    # Perform inference
    with torch.no_grad():
        outputs = protbert_model(tokens)

    # Access the output embeddings as needed
    embeddings = outputs.last_hidden_state

    # You can return relevant information based on your analysis needs
    return embeddings

```

```python
# Read the protein sequences from the FASTA file
Antibacterial = Antibacterial.dropna(subset=['Sequence'])

# Aplicar a função analyze_protein_sequence para cada sequência
Antibacterial['ProtBERT Embeddings'] = Antibacterial['Sequence'].apply(lambda seq: analyze_protein_sequence(seq, max_length=200))

# Imprimir o DataFrame resultante
print(Antibacterial)
```

```python
Defensis = Defensis.dropna(subset=['Sequence'])

# Aplicar a função analyze_protein_sequence para cada sequência
Defensis['ProtBERT Embeddings'] = Defensis['Sequence'].apply(lambda seq: analyze_protein_sequence(seq, max_length=200))

# Imprimir o DataFrame resultante
print(Defensis)
```
